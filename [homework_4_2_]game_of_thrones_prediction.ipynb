**Задание 1.4.** Создайте новые признаки.
* Создайте признак isPopular. У вас есть в таблице признак popularity score, поставьте какой то порог, например, 0.5, и пусть те персонажи, у которых popularity score меньше 0.5, непопулярны, т.е. isPopular = 0. Если же выше 0.5, тогда пусть isPopular=1.
* Создайте признак boolDeadRelations. Давайте упростим признак numDeadRelations, и просто поделим людей на тех, у кого были хоть какие то отношения с мертвыми персонажами, т.е. numDeadRelations > 0, и те, у которых не было, т.е. numDeadRelations = 0.
* Упростите признак culture, объединив схожие названия в один.


```python
df['popularity'] < 0.5
```

```python
df['isPopular'] = (df['popularity'] < 0.5).map({True: 0, False: 1})
df.drop(columns='popularity', inplace=True)

df_test['isPopular'] = (df_test['popularity'] < 0.5).map({True: 0, False: 1})
df_test.drop(columns='popularity', inplace=True)
```

```python
df['boolDeadRelations'] = (df['numDeadRelations'] >0).map({True: 1, False: 0})
df.drop(columns='numDeadRelations', inplace=True)

df_test['boolDeadRelations'] = (df_test['numDeadRelations'] >0).map({True: 1, False: 0})
df_test.drop(columns='numDeadRelations', inplace=True)
```

```python
columns_of_books = [f'book{num}' for num in range(1,6)]
df['num_appear']=df[columns_of_books].sum(axis=1)
df_test['num_appear']=df_test[columns_of_books].sum(axis=1)
```

```python
columns_of_siblings = ['mother', 'father', 'spouse', 'heir']

df['num_of_relatives']=df[columns_of_siblings].notna().sum(axis=1)
df_test['num_of_relatives']=df_test[columns_of_siblings].notna().sum(axis=1)
```

```python
#df.drop('has_heir')
df['has_heir'] = df['heir'].isna().map({True: 1, False: 0})
df_test['has_heir'] = df_test['heir'].isna().map({True: 1, False: 0})
```

```python
df.fillna({'dateOfBirth': 305 - df['age']}, inplace = True)
df_test.fillna({'dateOfBirth': 305 - df_test['age']}, inplace = True)
```

```python
# Подсказка
cult = {
    'Summer Islands': ['summer islands', 'summer islander', 'summer isles'],
    'Ghiscari': ['ghiscari', 'ghiscaricari',  'ghis'],
    'Asshai': ["asshai'i", 'asshai'],
    'Lysene': ['lysene', 'lyseni'],
    'Andal': ['andal', 'andals'],
    'Braavosi': ['braavosi', 'braavos'],
    'Dornish': ['dornishmen', 'dorne', 'dornish'],
    'Myrish': ['myr', 'myrish', 'myrmen'],
    'Westermen': ['westermen', 'westerman', 'westerlands'],
    'Westerosi': ['westeros', 'westerosi'],
    'Stormlander': ['stormlands', 'stormlander'],
    'Norvoshi': ['norvos', 'norvoshi'],
    'Northmen': ['the north', 'northmen'],
    'Free Folk': ['wildling', 'first men', 'free folk'],
    'Qartheen': ['qartheen', 'qarth'],
    'Reach': ['the reach', 'reach', 'reachmen'],
}
```

```python
for k, v in cult.items():
    df.loc[df['culture'].isin(v), 'culture']= k
df.fillna({'culture':'NoCulture'}, inplace = True)

for k, v in cult.items():
    df_test.loc[df_test['culture'].isin(v),'culture'] = k
df_test.fillna({'culture':'NoCulture'}, inplace = True)
```

```python
famous_houses = df['house'].value_counts().nlargest(5).index.to_list()
df.loc[(~(df['house'].isin(famous_houses)))& df['house'].notna(),'house'] = 'Rare'
df.fillna({'house':'NoInfo'}, inplace = True)

famous_houses = df_test['house'].value_counts().nlargest(5).index.to_list()
df_test.loc[(~(df_test['house'].isin(famous_houses)))& df_test['house'].notna(),'house'] = 'Rare'
df_test.fillna({'house': 'NoInfo'}, inplace = True)
```

```python
famous_titles = df['title'].value_counts().nlargest(5).index.to_list()
df.loc[(~(df['title'].isin(famous_titles)))& df['title'].notna(),'title'] = 'Rare'
df.fillna({'title':'NoTitle'}, inplace = True)

famous_titles = df_test['title'].value_counts().nlargest(5).index.to_list()
df_test.loc[(~(df_test['title'].isin(famous_titles)))& df_test['title'].notna(),'title'] = 'Rare'
df_test.fillna({'title':'NoTitle'}, inplace = True)
```

**Задание 1.10.** Разделите датасет на train и test часть при помощи функции `train_test_split`
